# New Delhi

```python
url="https://www.cars24.com/buy-used-car/?sort=P&storeCityId=2&pinId=110001"
new_delhi=[]
page=rq.get(url)
data=bs(page.text)
delhi=data.find_all("div",class_="_1l4fi")
for i in delhi:
    new_delhi.append(i.text)
#len(new_delhi)
# year
year_delhi=[]
for u in new_delhi:
    year_delhi.append(u.split(" ")[0])
#len(year_delhi)
# brand
brand_del=[]
for y in new_delhi:
    brand_del.append(y.split(" ")[1])
#len(brand_del)
# model
model_del=[]
for m in new_delhi:
    model_del.append(m.split(" ")[2])
#len(model_del)
# fuel type
fuel_del=[]
for f in new_delhi:
    fuel_del.append(re.findall(r"Petrol|Diesel",f)[0])
#len(fuel_del)
# km driven
km_del=[]
for k in new_delhi:
    km_del.append(re.findall(r"\w(\d+\W\d+)\skm|\w(\d+)\skm|(\d+\W\d+)\skm",k)[0])
#len(km_del)
#price
price_del=[]
for p in new_delhi:
    price_del.append(re.findall(r"\w\W(\d+\W\d+\W\d+)\w",p)[0])
#len(price_del)
# emi per month
emi_del=[]
for e in new_delhi:
    emi_del.append(re.findall(r"\W(\d+\W\d+)\Wmonth",e)[0])
#len(emi_del)
# transmission type
trm_del=[]
for t in new_delhi:
    trm_del.append(re.findall(r"Manual|Automatic|AUTOMATIC",t))
#len(trm_del)
# owner type
own_del=[]
for o in new_delhi:
    own_del.append(re.findall(r"1st|2nd",o))
#len(own_del)
Delhi=pd.DataFrame({"year":year_delhi,"brand":brand_del,"model":model_del,"fuel type":fuel_del,"km driven":km_del,"price":price_del,"emi/month":emi_del,"transmission type":trm_del,"owners":own_del})
Delhi["cities"]=pd.Series(["Delhi","Delhi","Delhi","Delhi","Delhi","Delhi","Delhi","Delhi","Delhi","Delhi","Delhi","Delhi","Delhi","Delhi","Delhi","Delhi","Delhi","Delhi","Delhi","Delhi"])
```
